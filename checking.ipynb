```python
import os
import numpy as np
from pdf2image import convert_from_path
import pickle

def convert_pdfs_to_arrays(pdf_file_dir, pdf_file_array_dir, dpi=300):
    """
    For each PDF file in 'pdf_file_dir', convert its pages to NumPy arrays,
    and save them in 'pdf_file_array_dir' as .npz files.
    
    :param pdf_file_dir: Directory containing PDF files.
    :param pdf_file_array_dir: Output directory to save the .npz files.
    :param dpi: Resolution for converting PDF pages to images (default=300).
    """
    # Ensure the output directory exists
    # os.makedirs(pdf_file_array_dir, exist_ok=True)

    # List all PDF files in the directory
    pdf_files = [
        f for f in os.listdir(pdf_file_dir)
        if f.lower().endswith('.pdf')
    ]


    num_files = len(pdf_files)

    for i, pdf_file in enumerate(pdf_files):
        pdf_path = os.path.join(pdf_file_dir, pdf_file)
        
        # Convert the PDF into a list of PIL images
        pages = convert_from_path(pdf_path, dpi=dpi)

        
        # Convert each PIL image to a NumPy array
        pages_array = [np.array(page) for page in pages]

        # Construct output filename (e.g., "filename.npz")
        base_name = os.path.splitext(pdf_file)[0]
        np_filename = f"{base_name}.np"
        np_path = os.path.join(pdf_file_array_dir, np_filename)

        # Save the list of page arrays into a compressed .npz file
        # Storing them as "pages" inside the archive
        # np.savez_compressed(np_path, pages=pages_array)
        with open(np_path,'wb') as f:
            pickle.dump(pages_array,f)
        
        print(f" {i+1} of {num_files} File(s) Saved: {np_filename} ({len(pages_array)} page(s)) in {pdf_file_array_dir}")

        # print(pages_array)
        # break
```

```python
import pyarrow.parquet as pq
import os
import pandas as pd

from tqdm import tqdm
# read all files
parquet_dir = "/root/autodl-tmp/data_1/lmms-lab/MP-DocVQA/data"
parquet_file_list = os.listdir(parquet_dir)
parquet_file_list = [p for p in parquet_file_list if p.find("test") != -1]
parquet_file_dict = [{"index":int(p.split("-")[1]),"file_name":p} for p in parquet_file_list]
parquet_file_dict = sorted(parquet_file_dict,key=lambda x: x['index'],reverse=False)
parquet_file_dict = [p['file_name'] for p in parquet_file_dict]

# get all images
data_all = []
for p in tqdm(parquet_file_dict):
    parquet_file = pq.ParquetFile(os.path.join(parquet_dir,p))
    data = parquet_file.read().to_pandas()
    data_all.append(data)

data_all = pd.concat(data_all,axis=0)
```
